```python
# FEATURES USED BY EACH TREE 

num = []
print(len(forest.estimators_))
i = 0
for tree_in_forest in forest.estimators_:
    H = set(tree_in_forest.tree_.feature)
    H.remove(-2)
    L = len(H) - 1
    num.append(L)
    if i%50 == 0:
        print(H)
    i=i+1
```

```python
# DECISION NODES USED BY EACH TREE 

dnodes = []
print(len(forest.estimators_))

for tree_in_forest in forest.estimators_:
    H = (tree_in_forest.tree_.feature)
    d = sum(H>=0)
    dnodes.append(d)
```

```python
#total questions asked:
sum(dnodes)
```

```python
# SAMPLES USED BY EACH TREE 

samples = []
print(len(forest.estimators_))
for tree_in_forest in forest.estimators_:
    H = (tree_in_forest.tree_.n_node_samples)
    d = H[0] #samples at root node
    samples.append(d)
```
