```python
import pandas as pd
```

```python
def update_importance(features_dict, kee, file_name, feat_imp, make_positive=False ):
    import pickle
    import copy

    
    # use 'key i.e. kee' to get the right dataframe
    df = features_dict[kee]
    
    # append a new column to the dataframe
    if(len(feat_imp) == 1):
        feat_imp = feat_imp[0]
    
    print('length of col vector : %d' % len(feat_imp))
    
    if (make_positive):
        print('making positive ...')
        pos = [ abs(x) for x in feat_imp ]
        feat_imp = pos
        
    df[file_name[:-7]] = copy.deepcopy(feat_imp)
    
    print(df.columns.values.tolist())
```

```python
feature_imp = {}
feature_imp['4c4']=pd.DataFrame()
feature_imp['4c3']=pd.DataFrame()
feature_imp['4c2']=pd.DataFrame()
```

```python
#write to disk
import pickle

with open('feature_importance.pkl', 'wb') as output:
    pickle.dump(feature_imp, output, pickle.HIGHEST_PROTOCOL)
```

```python
#load from disk
with open('feature_importance.pkl', 'rb') as input:
    features_dict = pickle.load(input)
```

```python
def fill_dict(files, kee):
    import pickle 
    global features_dict

    for fname in files:
        with open(fname, 'rb') as input:
            model = pickle.load(input)
            model_name = str(model)[:50]
            try:
                try:
                    x = len(model.feature_importances_)
                    print(model_name)
                    print('model.feature_importances_ with len = %d' % x)
                    #print(model.feature_importances_)
                    update_importance(features_dict, kee, fname, model.feature_importances_, False)
                except Exception as a:
                    #print('error ... %s' % a)
                    pass
            
                try:
                    y = len(model.coef_)
                    print(model_name)
                    print('model.coef_ with len = %d' % y)
                    #print(y)
                    update_importance(features_dict, kee, fname, model.coef_, True)
                except Exception as b:
                    #print('error...%s' % b)
                    pass
            except:
                pass
            print(60*'_')
```

```python
def plot_heatmap(matx, file_name, fsize=(10,20)):
    import seaborn as sns
    import matplotlib.pyplot as plt
    from sklearn.preprocessing import MinMaxScaler

    %matplotlib inline

    cols = matx.columns

    matx = MinMaxScaler().fit_transform(matx)
    matx = pd.DataFrame(matx, columns = cols)

    sns.set(context="paper", font="monospace")

    # Set up the matplotlib figure

    f, ax = plt.subplots(figsize=fsize)
    plt.setp(ax.get_xticklabels(), rotation='vertical', fontsize=20)
    plt.setp(ax.get_yticklabels(), fontsize=20)

    # Draw the heatmap using seaborn
    sns.heatmap(matx, vmax=1.0, square=False, linewidths=.5, center=0.5)
    plt.savefig(file_name)
    

    f.tight_layout()
```

```python
matx = features_dict['4c4']
plot_heatmap(matx, '4c4featureimportance.png')
```
